```python
import pandas as pd
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data')
```

```python
data.info() #type of features
```

```python
data.columns = ['edibility', 'x', 's', 'n', 't', 'odor', 'f', 'c', 'n.1', 'k', 'e', 'e.1', 's.1',
       's.2', 'w', 'w.1', 'p.2', 'w.2', 'o', 'p.3', 'k.1', 's.3', 'u']
```

```python
data
```

```python
data['edibility'] = data['edibility'].replace({'e': 'edible', 'p': 'poisonous'})
```

```python
data['odor'] = data['odor'].replace({'a': 'almond',
                                     'l': 'anise',
                                     'c': 'creosote',
                                     'y': 'fishy',
                                     'f': 'foul',
                                     'm': 'musty',
                                     'n': 'none',
                                     'p': 'pungent',
                                     's': 'spicy'
                                    })
```

```python
crosstab_odor_edibility = pd.crosstab(data['edibility'], data['odor'])
```

```python
crosstab_odor_edibility.columns[crosstab_odor_edibility.loc['poisonous', crosstab_odor_edibility.columns] > 0] #mushroom to be avoided while picking
```

```python
crosstab_odor_edibility['none']['edible']/(crosstab_odor_edibility['none']['edible']+crosstab_odor_edibility['none']['poisonous'])
```

```python
nutri = pd.read_excel('http://www.biostatisticien.eu/springeR/nutrition_elderly.xls')
```

```python
nutri = nutri.replace({
    'gender': {
        1: 'Male',
        2: 'Female'
    },
    'situation':{
        1: 'Single',
        2: 'Living with spouse',
        3: 'Living with Family',
        4: 'Living with someone else'
    },
    'meat': {
        0: 'Never',
        1: 'Less than once a week',
        2: 'Once a week',
        3: '2-3 times a week',
        4: '4-6 times a week',
        5: 'Everyday'
    },
    'fat': {
        1: 'Butter',
        2: 'Margarine',
        3: 'Peanut oil',
        4: 'Sunflower oil',
        5: 'Olive oil',
        6: 'Mix of vegetable oils',
        7: 'Colza oil',
        8: 'Duck or goose fat'
    }
})
```

```python
nutri.info()
```

```python
nutri.to_csv('nutrition_elderly.csv')
```
